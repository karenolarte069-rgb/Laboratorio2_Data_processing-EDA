```python
#1. Cargar el archivo JSON
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_json("pokemonDB_dataset.json")
#Convertir columnas a filas
df = pd.DataFrame(df).T
df
```

```python
#Explorar la estructura 
print(type(df))
print(df.index)
print(df.columns)
df.head(5)
```

```python
#2.Revisión de estructura y tipos 
df.info()                        # A este punto todas las columnas aparecen como categóricas porque los valores númericos están almacenados como texto en el archivo JSON original y eben convertirse manualmente
df.describe(include='all')       #Las variables no son númericas aún, las filas de la tabla corresponden a medidas descriptivas para variables categoricas
```

```python
#3.Limpieza y conversión de campos
#eliminación de unidades ("m","kg","lbs")
df["Height"].unique()
```

```python
df["Height"] = df["Height"].str.extract(r'([\d\.]+)').astype(float) #Extraer los valores númericos decimales de la columna "Height"
df.rename(columns={'Height': 'Height_m'}, inplace=True)
df.head(5)                                                          #r'' indica un raw string (cadena cruda), expresiones como \n,\t pasan a ser expresiones regulares y deben ser eliminadas 
                                                                    #\d representa cualquier dígito del 0 al 9 
                                                                    #\. es el punto decimal
                                                                    #+ indica "uno o más" de los caracteres entre corchetes
```

```python
df["Weight"].unique()
```

```python
df["Weight"] = df["Weight"].str.extract(r'([\d\.]+)').astype(float) #Extraer los valores númericos decimales de la columna "Weight"
df.rename(columns={'Weight': 'Weight_kg'}, inplace=True)
df.head(5)
```

```python
df["Egg Cycles"].unique()
```

```python
df["Egg Cycles"] = df["Egg Cycles"].str.extract(r'(\d+)').astype('Int64')
df["Egg Cycles"].head(5)
```

```python
#Convertir columnas categoricas a volores númericos
Col_num=[ 
    "Height", "Weight",
    "HP Base", "HP Min", "HP Max",
    "Attack Base", "Attack Min", "Attack Max",
    "Defense Base", "Defense Min", "Defense Max",
    "Special Attack Base", "Special Attack Min", "Special Attack Max",
    "Special Defense Base", "Special Defense Min", "Special Defense Max",
    "Speed Base", "Speed Min", "Speed Max",
    "Base Exp", "Egg Cycles"
    ]

for Col in Col_num:
    if Col in df.columns:
        df[Col] = pd.to_numeric(df[Col], errors='coerce')    # errors='coerce' convierte los valores no númericos o vacios a NaN 
```

```python
df.info()
```

```python
df[["Type1", "Type2"]] = df["Type"].str.split(",", n=1, expand=True) #Definir nuevas columnas a reemplazar "Type"
                                                                     #.Split divide el texto utilizando la primera coma(separador)
                                                                     #Expand=True panda expande los elementos de la lista en columnas separadas del df

df["Type1"] = df["Type1"].str.strip()  #.Strip=Elimina espacios
df["Type2"] = df["Type2"].str.strip()

pos = df.columns.get_loc("Type") #Posición original de la columna "Type"

df.drop(columns=["Type"], inplace=True) #Eliminar la columna "Type"
 
cols = list(df.columns)                 #Listas de columnas del df
cols = cols[:pos] + ["Type1", "Type2"] + cols[pos:-2]  #cols[:pos] = Toma las columnas antes de Type, cols[pos:-2] = Toma las columnas despues de Type sin incluir las dos últimas
df = df[cols]
```

```python
df["Abilities"] = df["Abilities"].str.replace(r"\d+\.\s*", "", regex=True)
```

```python
df[["Ability1", "Ability2", "Ability3"]] = df["Abilities"].str.split(",",n=2, expand=True) 

df["Ability1"] = df["Ability1"].str.strip()  
df["Ability2"] = df["Ability2"].str.strip()
df["Ability3"] = df["Ability3"].str.strip()

pos = df.columns.get_loc("Abilities") 

df.drop(columns=["Abilities"], inplace=True) 

cols = list(df.columns)                
cols = cols[:pos] + ["Ability1", "Ability2", "Ability3"] + cols[pos:-3]  
df = df[cols]
```

```python
df[["Male(%)", "Female(%)"]] = df["Gender"].str.split(",", n=1, expand=True) 

df["Male(%)"] = df["Male(%)"].str.strip()  #.Strip=Elimina espacios
df["Female(%)"] = df["Female(%)"].str.strip()

pos = df.columns.get_loc("Gender") #Posición original de la columna "Gender"

df.drop(columns=["Gender"], inplace=True) #Eliminar la columna "Gender"
 
cols = list(df.columns)                 #Listas de columnas del df
cols = cols[:pos] + ["Male(%)", "Female(%)"] + cols[pos:-2]  #cols[:pos] = Toma las columnas antes de Gender, cols[pos:-2] = Toma las columnas despues de Gender sin incluir las dos últimas
df = df[cols]
```

```python
df.info()
```

```python
#4.Normalización y verificación final 
#Verificación de valores nulos:
print(df.isnull().sum())
```

```python
#DataFrame limpio para el analisis:
df_clean=df[["Type1", "Type2", "HP Base", "Attack Base", "Defense Base", "Speed Base", "Height_m", "Weight_kg", "Base Exp"]]
```

```python
df_clean = df_clean.dropna()
df_clean.info()
df_clean.head(5)
```

```python
df_clean.describe().round(2)
```

Etapa 2: Exploratory Data Analysis (EDA)
1. Análisis estadístico 
¿Cuál es el promedio, mínimo y máximo de los atributos base (HP, 
Attack, Defense, Speed) de todos los Pokémon? 

```python
cols = ["HP Base", "Attack Base", "Defense Base", "Speed Base"]
# Calcular promedio, mínimo y máximo
resumen = pd.DataFrame({
    "Promedio": df_clean[cols].mean(),
    "Mínimo": df_clean[cols].min(),
    "Máximo": df_clean[cols].max()
})
resumen = resumen.round(2)

print(resumen)
```

En "HP Base" o puntos de vida se observa un promedio de 73 puntos, con valor mínimo de 1 para los Pokémon mas débiles y un máximo de 223 para los que poseen mayor capacidad de resistencia. 

Dado que el promedio de "Attack Base" es superior en comparación al promedio de "Defense Base", los pokémon tienen mayor capacidad de ataque que defensa.Sin embargo, el valor máximo de 230  de defensa demuestra el contraste de ciertos pokémon con una resistencia extrema.

El promedio de "Speed Base" demuestra que los pokémon no se caracterizan por su velocidad ya que la mayoría se concentran en rangos medios o bajos.

Genere un gráfico de barras que muestre la cantidad de Pokémon por

6. Análisis estadístico 
Calcule la correlación entre los atributos HP Base, Attack Base, 
Defense Base y Speed Base. 
¿Qué atributos están más correlacionados entre sí?

```python
atributos = ["HP Base", "Attack Base", "Defense Base", "Speed Base"]

# Calcular la matriz de correlación
correlacion = df_clean[atributos].corr()

print("Matriz de correlación:")
print(correlacion)
```

Teniendo en cuenta que un coeficiente de correlación de 1 o cercano a 1 indica una correlación positiva fuerte, los atributos más correlacionados entre sí son "Hp Base" y "Attack Base" (r=0.50) seguida de "Attack Base" y "Defense Base" (r=0.42), ambas siendo correlación positiva moderada. 

La relación entre "Speed Base" y "HP Base" (r = 0.19) es muy débil, y entre "Defense Base" y "Speed Base" (r = -0.06) es negativa débil o nula.

7. Análisis gráfico 
Cree un heatmap (mapa de calor) con la matriz de correlaciones 
obtenida en el punto anterior. 
Interprete los resultados. 

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlacion, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Mapa de calor de correlaciones entre atributos base")
plt.show()
```

La matriz de correlación muestra que existe una relación positiva moderada entre los atributos "Attack Base" y "Defense Base", lo que indica que los Pokémon con un mayor nivel de ataque tienden también a tener una defensa más alta. De manera similar, se observa una correlación moderada entre "HP Base" y "Attack Base", lo que sugiere que los Pokémon con más puntos de vida suelen poseer también un ataque superior.

8. Análisis gráfico 
Realice un diagrama de dispersión (scatter plot) entre Weight_kg y 
Attack Base. 
¿Existe relación entre el peso de un Pokémon y su capacidad de ataque?

```python

plt.figure(figsize=(10,6))
plt.scatter(df_clean["Weight_kg"], df_clean["Attack Base"], color="red", alpha=0.6, s=15)

plt.xlabel("Peso (kg)", fontsize=12)
plt.ylabel("Ataque base", fontsize=12)
plt.title("Relación entre el peso y la capacidad de ataque de los Pokémon", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

A medida que el peso aumenta, los puntos se dispersan, esto indica que no existe una relación lineal evidente entre peso y ataque de base, por lo cuál, mayor peso no es indicativo de ataque de base más alto. 

9. Análisis estadístico 
Determine el promedio de altura y peso por tipo principal (Type1). 
Interprete cuál tipo tiende a tener Pokémon más grandes.

```python

# Calcular promedios por Type1
promedios = df_clean.groupby("Type1")[["Height_m", "Weight_kg"]].mean().sort_values("Height_m", ascending=False)

# Mostrar resultados
print("Promedio de altura y peso por tipo principal:\n")
print(promedios)

print("\nTipo con Pokémon más altos:", promedios["Height_m"].idxmax())
print("Tipo con Pokémon más pesados:", promedios["Weight_kg"].idxmax())
```

Los Pokémon de tipo Steel son, en promedio, los más altos y los más pesados entre todos los tipos principales.
Esto se debe a que en este grupo se encuentran especies con cuerpos metálicos o armaduras, lo que incrementa significativamente tanto su altura como su peso promedio en comparación con otros tipos.

El tipo "Dragon" también presenta valores elevados tanto ppara altura como para el peso. Los tipo "Normal", "Electric" y "Bug" son más bajos y tienen menor peso.


10. Análisis gráfico 
Construya un gráfico de violín o boxplot múltiple comparando el 
atributo Speed Base entre los tipos Flying, Electric y Ground. 
¿Qué tipo de Pokémon tiende a ser más rápido? 

```python


# Filtrar los tipos que se van a comparar
tipos = ["Flying", "Electric", "Ground"]
df_tipos = df_clean[df_clean["Type1"].isin(tipos)]

plt.figure(figsize=(8,6))
sns.violinplot(data=df_tipos, x="Type1", y="Speed Base", hue="Type1", palette="Set2", inner="quartile", legend=False)

plt.title("Distribución de la velocidad base por tipo", fontsize=14, fontweight="bold")
plt.xlabel("Tipo de Pokémon", fontsize=12)
plt.ylabel("Velocidad base", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

Los Pokémon de tipo Electric tienden a ser los más rápidos, ya que presentan las velocidades base más altas y una mayor concentración de valores elevados.
