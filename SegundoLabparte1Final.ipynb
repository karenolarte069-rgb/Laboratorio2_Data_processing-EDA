```python
import pandas as pd
df= pd.read_csv("census.csv")
```

```python
# Veamos el dataset:
df
```

```python
# Descripción de las columnas:
df.info()
df.shape[1]                 #Dimensiones del df, (0,1), 1-->columnas, #de columnas=8
```

```python
# Verifiquemos si hay datos faltantes:
print('Valores faltantes por columna:\n ')
print(df.isnull().sum(), '\n') #Devuelte un df con las mismas diemsio con True si el dato esta vacio (NaN) o false si no lo esta, . sum()suma los valores True e cada columna 
```

```python
# Descartemos las filas que tengan 3 o más datos faltantes:
df=df.dropna(thresh= df.shape[1] - 3)        #.dropna elimina las filas o columnas con valores faltantes(NaN)
                                             #df.shape arroja las dimensiones del dataframe (0,1), 1-->columnas, #de columnas=8
                                             #thresh=numero minimo de valores validos no nulos, en este caso sería, 8-3=5, 5 valores minimos no nulos
```

```python
# Contemos de nuevo los datos faltantes:
print('Valores faltantes tras descartar filas con tres o más valores faltantes: \n')
print(df.isnull().sum(), '\n')
```

```python
# Imputemos los datos faltantes de edad y horas trabajadas por semana con la mediana de cada una de esas columnas:
for i in ['age', 'hours_per_week']:
    df.loc[:, i] = df[i].fillna(df[i].median())                
                                                               
print('Valores faltantes tras imputar con la mediana: \n ')    
print(df[['age', 'hours_per_week']].isnull().sum(), '\n')      
```

```python
# Contemos de nuevo los datos faltantes:
print(df.isnull().sum(), '\n')
```

```python
# Apliquemos one-hot encoding a la columna "workclass":
df=pd.get_dummies(df, columns=['workclass'], dtype=int)
                  
                  
                  
```

```python
# Hallemos los valores que toma la columna "education":
df=df.sort_values(by='education')
df['education'].unique()
```

```python
# Apliquemos ordinal encoding a la columna "education":
from sklearn.preprocessing import OrdinalEncoder 
encoder = OrdinalEncoder()                       
df['education_encoded'] = encoder.fit_transform(df[['education']]) 
                                                                   
                                                                   
                                                                   
                                                                  
                                                                   
df
```

```python
# Verifiquemos que la columna "education" tenga los valores apropiados:
df['education'].unique()
```

```python
# Apliquemos one-hot encoding a la columna "race":
df=pd.get_dummies(df, columns=['race'], dtype=int)
df
```

```python
# Apliquemos binary encoding a la columna "sex":
df['sex_binary']=df['sex'].map({'Female': 0, 'Male': 1})
                                                        
                                                        
df
```

```python
# Apliquemos binary encoding a la etiqueta:
df['label_binary']=df['label'].map({'<=50K':0, '>50K':1})
df
```
